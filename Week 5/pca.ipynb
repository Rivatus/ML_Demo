```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat = scipy.io.loadmat('small_data.mat')
```

```python
mat.items()
```

```python
X = np.array(mat["X"])
X
```

```python
X_norm = (X - np.mean(X, axis = 0)) / np.std(X, axis = 0)
X_norm
```

```python
plt.plot(X_norm[:, 0], X_norm[:, 1], "bo")
```

```python
Sigma = 1/50 * np.dot(X_norm.T, X_norm)
E, S = np.linalg.eig(Sigma)
#print(U[0, 0], U[1, 0])
S
```

```python
K = 1
Z = np.dot(X_norm, S[:, 0:K])
Z
```

```python
X_rec = np.dot(Z, S[:, 0:K].T)
X_rec
```

```python
# Plot the actual data points and the recovered data points

fig, ax = plt.subplots(figsize = (8,8))
ax.plot(X_rec[:,0], X_rec[:,1], "ro")
ax.plot(X_norm[:,0], X_norm[:,1], "bo")

for i in range(X_norm.shape[0]):
    xs = [X_rec[i,0], X_norm[i,0]]
    ys = [X_rec[i,1], X_norm[i,1]]
    ax.plot(xs, ys, "g--")
    
ax.set_aspect("equal")
```

## Eigenfaces

```python
import pandas as pd
```

```python
df = pd.read_csv("face_data.csv")
df.head()
```

```python
df.shape
```

```python
df["target"].nunique()
```

```python
def plot_faces(pixels):
    fig, axes = plt.subplots(5, 5, figsize = (8, 8))
    for i,ax in enumerate(axes.flat):
        img = np.array(pixels[i]).reshape(64, 64)
        ax.imshow(img, cmap = "gray")
```

```python
X = df.drop("target", axis=1)
Y = df["target"]
X = X.to_numpy()
X.shape
```

```python
plot_faces(X)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
X_pca = pca.fit_transform(X)
X_pca
```

```python
plt.bar(range(400), pca.explained_variance_)
```

```python
plt.plot(pca.explained_variance_ratio_.cumsum())
```

```python
np.where(pca.explained_variance_ratio_.cumsum() >= 0.95)
```

```python
pca_final = PCA(n_components = 122)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.5)
Y_train.shape
```

```python
X_train_pca = pca_final.fit_transform(X_train)

X_pca_rec = pca_final.inverse_transform(X_train_pca)
X_pca_rec.shape
```

```python
plot_faces(X_pca_rec)
```

```python
from sklearn.neural_network import MLPClassifier

classifier = MLPClassifier(hidden_layer_sizes = (1024,), activation = "relu", batch_size = 64, verbose = True, early_stopping = True).fit(X_train_pca, Y_train)
```

```python
X_test_pca = pca_final.transform(X_test)

classifier.score(X_test_pca, Y_test)
```
